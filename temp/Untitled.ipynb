```python
from bokeh.io import show, output_notebook
from bokeh.plotting import figure
from datetime import datetime

output_notebook()
```

```python
deci_data, deci_ts = [], []
full_data, full_ts = [], []

with open("dump_deci_001.dat", "r") as f:
    f.readline()
    f.readline()
    for l in f.readlines():
        spl = l.split(',')
        ts = datetime.strptime(spl[0][:-3], "%Y-%m-%d_%H:%M:%S.%f")
        deci_ts.append(ts)
        deci_data.append(float(spl[1][2:]))

with open("dump_full_001.dat", "r") as f:
    f.readline()
    f.readline()
    for l in f.readlines():
        spl = l.split(',')
        ts = datetime.strptime(spl[0][:-3], "%Y-%m-%d_%H:%M:%S.%f")
        full_ts.append(ts)
        full_data.append(float(spl[1][2:]))
```

```python
fig = figure(width=1200, height=600, x_axis_type = 'datetime')
fig.circle(deci_ts, deci_data, color='red', legend_label='Deci', size=8)
fig.cross(full_ts, full_data, color='blue', legend_label='Full')

show(fig)
```
